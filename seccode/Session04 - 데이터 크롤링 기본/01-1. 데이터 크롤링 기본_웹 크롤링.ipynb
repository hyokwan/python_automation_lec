# 3. 태그정보 수집하기

### (1) 태그명으로 소스 수집하기

```python
# 코드 내 a 태그 1개만 탐색
bs = bs4.BeautifulSoup(html,"html.parser")
aTag = bs.find(name = "a")
print(aTag)
print( type(aTag) )
# 코드 내 a 태그 탐색, 상단 2개만
aTags = bs.findAll(name = "a", limit = 2)
print(aTags)
print( type(aTags) )
```

# 4. 테이블 정보 수집하기

### 테이블트리 스크랩

```python
import requests, bs4
import pandas as pd

#웹페이지 html 소스 가져오기
resp = requests.get("https://en.wikipedia.org/wiki/Help:Table")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')

tabletag = bs.find("table", {"class":"wikitable", "style":"height:14em"})
tabletag
```

```python
rows = tabletag.find("tbody").findAll("tr")
rows
```

```python
# 행 전체를 저장할 리스트
rowList=[]
# 행별 컬럼값을 저장할 리스트
columnList =[]
```

```python
for i in range(0, len(rows)):
    columns = rows[i].findAll(["th","td"])
    columnsLen = len(columns)
    for j in range(0, columnsLen):
        columnList.append(columns[j].text.replace("\n",""))

    rowList.append(columnList)
    columnList=[]
```

```python
rowList
```

```python
df = pd.DataFrame(rowList)
df.columns = df.iloc[0] 
df = df.iloc[1:]
df.head()
```
