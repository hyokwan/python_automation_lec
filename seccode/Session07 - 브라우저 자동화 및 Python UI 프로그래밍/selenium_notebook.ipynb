```python
import bs4

# 라이브러리 선언
from selenium import webdriver
import bs4
# 드라이버 위치 설정
driver_loc = "../externlib/chromedriver/chromedriver.exe"

# 드라이버 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("window-size=1920x1080")

# 헤드리스 설정 시 리눅스 같이 웹화면 없는경우에도 웹페이지 실행가능함
options.add_argument("headless")
options.add_argument("disable-gpu")
# 웹 드라이버 정의
driver = webdriver.Chrome(driver_loc, options=options)
# 웹페이지 파싱 될때까지 최대 3초 기다림
driver.implicitly_wait(3)
```

```python
from selenium.webdriver.common.keys import Keys

# URL 정의
googleUrl = 'https://www.google.com'

# URL 이동
driver.get(googleUrl)

# 구글탐색키 입력
searchPath = '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input'

# time.sleep(3)
driver.find_element_by_xpath(searchPath).send_keys("selenium wiki")
# driver.execute_script("arguments[0].click();", element)
# time.sleep(3)
driver.find_element_by_xpath(searchPath).send_keys(Keys.ESCAPE)
# time.sleep(3)
buttonPath = '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[3]/center/input[1]'
driver.find_element_by_xpath(buttonPath).click()
```
