### 4. 데이터 기초통계량 분석

```python
# 전체데이터 기초 상관분석
scoreData.corr()
```

```python
sns.heatmap(scoreData.corr())
plt.title('Covariance Plot')
```

```python
lectureColumns = scoreData.loc[0:1,"LEC01":"LEC04"].columns
```

##### 4-1 과목별 상관관계 분석

```python
count = 1
#Scatter plot of scores in different subjects to visualize correlation among them
plt.figure(figsize = (20,20))
for i in range(0, len(lectureColumns)):
    for j in range(0, len(lectureColumns)):
        plt.subplot(4,4,count)
        plt.scatter(scoreData[lectureColumns[i]],scoreData[lectureColumns[j]])
#         plt.title(" {} vs {} ".format(lectureColumns[i],lectureColumns[j] ) )
        plt.xlabel(lectureColumns[i])
        plt.ylabel(lectureColumns[j])
        count = count + 1
```

##### 4-2 과목별 성적 분포도

```python
plt.hist([scoreData['LEC01'],scoreData['LEC02'],scoreData['LEC03'],scoreData['LEC04']], color=['red', 'yellow', 'blue', "green"])
plt.title('score hist')
plt.xlabel('score')
plt.ylabel('Number of Students')
plt.legend(lectureColumns)
plt.show()
```

##### 4-3 과목별 평균

```python
lectureLen = len(lectureColumns)
```

```python
y = lectureColumns
width = [scoreData.loc[:,lectureColumns[i]].mean() for i in range(0, lectureLen)]

plt.figure(figsize = (12,2))
plt.barh(y = y, 
         width = width)
plt.title('Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Lectures')
for i,v in enumerate(width):
    plt.text(v, i, " "+str(round(v,2)), color='blue', va='center', fontweight='bold')
plt.show()
```
