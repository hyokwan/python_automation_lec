# 1. 사고이력 데이터 전처리

### 1-1. 데이터 병합 (사고이력, 선박마스터정보)

```python
# 사고이력 데이터 불러오기
sisData = pd.read_csv\
        ("../dataset_h/sis_dat_fin.csv", encoding="ms949")
# 선박 마스터 정보 불러오기
shipMaster= pd.read_csv\
        ("../dataset_h/SHIP_MASTER.csv", encoding="ms949")
```

```python
shipMaster.head(2)
```

```python
sisData = sisData.dropna(axis=1, how="all")
shipMaster = shipMaster.dropna(axis=1, how="all")
```

```python
# 마스터 정보에서 가져올 SHIP_GUBUN 코드 선택 (1: 어선, 2: 비어선)
masterCoreColumns = ["SHIP_NO","SHIP_GUBUN"]
shipMaster = shipMaster[masterCoreColumns]
```

```python
shipMaster.head(1)
```

```python
# 선박번호 14자리 설정 어선 중심으로 무조건 14자리 맞추고 마스터 정보와 병합
sisData.선박번호 = sisData.선박번호.astype(str)
shipMaster.SHIP_NO = shipMaster.SHIP_NO.astype(str)
# sisData["선박번호_기존"] = sisData["선박번호"]
sisData["선박번호"] = sisData["선박번호"].astype(str).str.pad(width=14, side='left', fillchar='0')
```

```python
# 데이터 병합 (사고데이터, 마스터 데이터)
sisMergedData = pd.merge(left=sisData,
         right=shipMaster,
         left_on="선박번호",
         right_on="SHIP_NO",
         how="inner")
```

```python
sisMergedData.shape
```

### 2-1. 정기검사 병합 (CHECKLIST2.csv)

```python
checkList2 = pd.read_csv\
        ("../dataset_h/check_list2_new.csv", encoding="ms949")

print(checkList2.shape)

# 전체 컬럼값 NULL 컬럼삭제
checkList2Refined = checkList2.dropna(axis=1, how="all")

print(checkList2Refined.shape)

# 검사지사, 접수년도, 검사종률, 접수번호
joinKey = ["OFFICE_GUBUN","ACCT_YEAR","SHIP_KIND", "ACCT_NO"]

### 데이터 테스트

kitCheck2 = pd.merge(left=kitProcRefined,
         right=checkList2Refined,
         on=joinKey,
         suffixes=('', '_C2'),
         how="inner")

print(kitCheck2.shape)

kitCheck2.dropna(axis=1, how="all", inplace=True)
print(kitCheck2.shape)

print(len(kitCheck2.SHIP_NO.drop_duplicates()))

# # null 컬럼 90% 초과 컬럼 제거
# limitColumnValue = int(kitCheck2.shape[0]*NULL_PER)
# nullCheck = pd.DataFrame( kitCheck2.isnull().sum() , columns = ["NULL_CNT"])
# dropColumns = list(nullCheck.loc[nullCheck.NULL_CNT > limitColumnValue].index)

# kitCheck2.loc[~kitCheck2.CHECK_JUNG.isnull()].head(2)

# kitCheck2.loc[kitCheck2.SHIP_NO=="96030486458003"].head(2)

sisCheckData2 = pd.merge(left=sisRefinedData,
         right=kitCheck2,
         left_on="선박번호",
         right_on="SHIP_NO",
         suffixes=('', '_CK20'),
         how="inner")

print(sisCheckData2.shape)

print(len(sisCheckData2.SHIP_NO.drop_duplicates()))
```

```python
analData = sisCheckData2.loc[~sisCheckData2.해양사고발생_년도.isnull()]

# analData[tColumns].isnull().sum()

analData2 = analData.loc[~analData.VERIFY_T_DATE.isnull()]

# analData2[tColumns].isnull().sum()

analData2.선박번호.drop_duplicates()

analData2["ACC_DATE"] = analData2["해양사고발생_년도"].astype(int).astype(str)+\
                        analData2["해양사고발생_월"].astype(int).astype(str).str.pad(width=2, side='left', fillchar='0')+\
                        analData2["해양사고발생_일"].astype(int).astype(str).str.pad(width=2, side='left', fillchar='0')

analData2.loc[analData2.ACC_DATE.str.len() != 8].head()

def dateConvert(inValue):
#     inValue = "19/12/30"
    listData = inValue.split("/")

    yearValue = listData[0]
    monthValue = listData[1].zfill(2)
    dateValue = listData[2].zfill(2)

    if int(yearValue) > 50:
        yearValue = "19"+yearValue
    else:
        yearValue = "20"+yearValue

    outValue = yearValue+monthValue+dateValue
    return outValue

analData2.loc[analData2.VERIFY_F_DATE.str.len() != 8]["VERIFY_F_DATE"]

analData2["CERTIFI_FR_DATE2"]= analData2.VERIFY_F_DATE.apply(dateConvert)

analData2["CERTIFI_TO_DATE2"]= analData2.VERIFY_T_DATE.apply(dateConvert)

analData2.loc[analData2.CERTIFI_FR_DATE2.str.len() != 8].head()

analDataFinal = analData2.loc[ (analData2.ACC_DATE >= analData2.CERTIFI_FR_DATE2) & \
                               (analData2.ACC_DATE <= analData2.CERTIFI_TO_DATE2) ]

analDataFinal.shape

len(analDataFinal.SHIP_NO.drop_duplicates())

sortKey = ["선박번호","ACC_DATE"]

analDataFinal = analDataFinal.sort_values(sortKey)
```

```python
analData = sisCheckData21.loc[~sisCheckData21.해양사고발생_년도.isnull()]

# analData[tColumns].isnull().sum()

analData2 = analData.loc[~analData.VERIFY_T_DATE.isnull()]

# analData2[tColumns].isnull().sum()

analData2.선박번호.drop_duplicates()

analData2["ACC_DATE"] = analData2["해양사고발생_년도"].astype(int).astype(str)+\
                        analData2["해양사고발생_월"].astype(int).astype(str).str.pad(width=2, side='left', fillchar='0')+\
                        analData2["해양사고발생_일"].astype(int).astype(str).str.pad(width=2, side='left', fillchar='0')

analData2.loc[analData2.ACC_DATE.str.len() != 8].head()

def dateConvert(inValue):
#     inValue = "19/12/30"
    listData = inValue.split("/")

    yearValue = listData[0]
    monthValue = listData[1].zfill(2)
    dateValue = listData[2].zfill(2)

    if int(yearValue) > 50:
        yearValue = "19"+yearValue
    else:
        yearValue = "20"+yearValue

    outValue = yearValue+monthValue+dateValue
    return outValue

analData2.loc[analData2.VERIFY_F_DATE.str.len() != 8]["VERIFY_F_DATE"]

analData2["CERTIFI_FR_DATE2"]= analData2.VERIFY_F_DATE.apply(dateConvert)

analData2["CERTIFI_TO_DATE2"]= analData2.VERIFY_T_DATE.apply(dateConvert)

analData2.loc[analData2.CERTIFI_FR_DATE2.str.len() != 8].head()

analDataFinal21 = analData2.loc[ (analData2.ACC_DATE >= analData2.CERTIFI_FR_DATE2) & \
                               (analData2.ACC_DATE <= analData2.CERTIFI_TO_DATE2) ]

print(analDataFinal21.shape)

print(len(analDataFinal21.SHIP_NO.drop_duplicates()))

sortKey = ["선박번호","ACC_DATE"]

analDataFinal21 = analDataFinal21.sort_values(sortKey)
```
